```python
#DATA PREPROCESSING

data["Gender"].fillna(data["Gender"].mode()[0],inplace=True)
data
```

```python
data["Gender"]=data["Gender"].astype(int)
data
```

```python
data["DailyTimeSpentonSite"]=data["DailyTimeSpentonSite"].fillna(data["DailyTimeSpentonSite"].mean())
data
```

```python
data["AreaIncome"]=data["AreaIncome"].fillna(data["AreaIncome"].median())
data
```

```python
data["DailyInternetUsage"]=data["DailyInternetUsage"].fillna(data["DailyInternetUsage"].mean())
data
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data["Country"]=le.fit_transform(data["Country"])
data
```

```python
def timeconv(value):
    hours=int(value//60)
    minutes=int(value%60)
    return f"{int(hours)}:{int(minutes)}"
data["DailyTimeSpentonSite"]=data["DailyTimeSpentonSite"].apply(timeconv)
print(data["DailyTimeSpentonSite"])
```

```python
data=data.drop(columns=["AdTopicLine","City"])
data
```

```python
data[["DataSpendInHours","DataSpendInMinutes"]]=data["DailyTimeSpentonSite"].str.split(":",expand=True)
data["DataSpendInHours"]=data["DataSpendInHours"].astype(int)
data["DataSpendInMinutes"]=data["DataSpendInMinutes"].astype(int)
data
```

```python
data[["Date","Time"]]=data["Timestamp"].str.split(" ",expand=True)
data=data.drop("Timestamp", axis=1)
data=data.drop("DailyTimeSpentonSite",axis=1)
data
```

```python
#new dataset
df=data
df
```

```python
#Logistic Regression Model
from sklearn.model_selection import train_test_split
x = df[['DataSpendInHours', 'DataSpendInMinutes', 'Age', 'AreaIncome','DailyInternetUsage', 'Gender']]
y = df['ClickedonAd']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=101)
```

```python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
x.isnull().sum()
```

```python
#LOGISTIC REGRESSOR
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
y_pred =model.predict(X_test)
print(y_pred)
```

```python
dff=pd.DataFrame({"Actaul y value": y_test, "Predicted y value": y_pred})
dff.tail()
```

```python
#Training score
model.score(X_train,y_train)
```

```python
#Testing score
model.score(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print("Accuracy:" ,accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```
